## Census Data Discovery
### Urban Renewal Explorer - Exploratory Analysis

A main component of my final project is analyzing socioeconomic change in neighborhoods that experienced urban renewal over several decades. The challenge is that Census tract boundaries change regularly which makes direct comparison over time more complicated than simply lining up decennial Census data.

Based on online research, and a mention in MUSA 5080, the [Longitudinal Tract Data Base](https://s4.ad.brown.edu/projects/diversity/Researcher/Bridging.htm) seemed like a promising contender to address this challenge. This database provides estimates within 2010 tract boundaries for tract-level census data, going back as early as 1970. Of course, this excludes some of the immediate impacts of Urban Renewal in the 50s and 60s, and makes it harder to interpret 2020 data (outisde of race). This can be explored in future improvements to this project.
